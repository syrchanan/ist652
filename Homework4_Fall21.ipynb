# Code for data analysis 1

You can place the code for your first data analysis result in this section. Add as many code cells as you need.

```python
crimes_description = crimes.set_index(['Description'], append=True) 
crimes_description #add index
```

```python
crimes_description['times'] = np.ones(len(crimes_description), dtype=int) #trickiness to be able to combine since there are
crimes_combined = crimes_description.times.groupby(level=[0,1]).sum() #duplicates for the same hours and days
```

```python
crimes_combined[crimes_combined>1].count() #check that it works
```

```python
crimes_by_desc = crimes_combined.unstack(level=1).fillna(0).resample('h').sum().fillna(0).astype(int) #summarise the stat as a sum
```

```python
crimes_by_desc #view
```

```python
print(crimes_by_desc.columns.tolist()) #check options for the columns
```

```python
crimes_by_desc.loc[:,["ILLEGAL CONSUMPTION BY MINOR", "ILLEGAL POSSESSION BY MINOR"]].resample('Y').sum().plot()
#subset and plot desired columns over time
```

## Description of data analysis result 1

Use the next cell to describe your data analysis result 1

Over the course of 4 years, the total number of Illegal Consumption and Illegal Possession charges for minors has decreased. There was not a huge number to begin with, but there has been a definite decline over time. For example, Illegal Possession cases decreased from over 20 in 2018 to just a few in 2021 (to date).

```python
crimes_arrest_bool = crimes.set_index(['Arrest'], append=True) 
crimes_arrest_bool #add index
```

```python
crimes_arrest_bool['counts'] = np.ones(len(crimes_arrest_bool), dtype=int) #trickiness to be able to combine since there are
crimes_bool_combo = crimes_arrest_bool.counts.groupby(level=[0,1]).sum() #duplicates for the same hours and days
```

```python
crimes_bool_combo[crimes_bool_combo>1].count() #check that it works
```

```python
crimes_bool_desc = crimes_bool_combo.unstack(level=1).fillna(0).resample('h').sum().fillna(0).astype(int) #summarise the stat as a sum
```

```python
crimes_bool_desc.resample('M').sum().plot()
#subset and plot desired columns over time
```
